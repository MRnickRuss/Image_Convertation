```python
from imageai.Classification import ImageClassification
import os
```

```python
import numpy as np
import cv2
from google.colab.patches import cv2_imshow
from urllib.request import urlopen
req = urlopen('https://www.psypage.ru/wp-content/uploads/2017/10/25749_shutterstock_96240233.jpg')

image_2 = np.asarray(bytearray(req.read()), dtype=np.uint8)
image_2 = cv2.imdecode(image_2, -1)

cv2_imshow(image_2)
```

```python
execution_path = "inception_v3_google-1a9a5a14.pth"

prediction = ImageClassification()
prediction.setModelTypeAsInceptionV3()
prediction.setModelPath(execution_path)
prediction.loadModel()
```

```python
predictions, probabilities = prediction.classifyImage((image_2), result_count=10)
for eachPrediction, eachProbability in zip(predictions, probabilities):
    print(eachPrediction , " : " , eachProbability)
```
